```python
def ackley_function(x, a=20, b=0.2, c=2*np.pi, axis=1):
    x = np.asarray(x)
    factor1 = np.sqrt(1. * np.square(x).sum(axis=axis) / x.shape[axis])
    factor1 = -a * np.exp(-b * factor1)

    factor2 = np.exp(1. * np.cos(c * x).sum(axis=axis) / x.shape[axis])

    return factor1 - factor2 + a + np.e
```

```python
def shubert_function(x, param=418.9829, axis=1):
    x = np.asarray(x)
    sqrt_abs = np.sqrt(np.abs(x))
    sum_sin_sqrt = (x * np.sin(sqrt_abs)).sum(axis=axis)
    return param * x.shape[axis] - sum_sin_sqrt
```

```python
def function(x, axis=1):
    x = np.asarray(x)
    dim = x.shape[axis]
    res = 0.

    square_x = np.square(x)
    square_x_1 = np.square(x-1)

    term1 = 100. * np.square(x[:, 1:] - square_x[:, :-1])
    return (term1 + square_x_1[:, :-1]).sum(axis=axis)
```
